```python
import os
import pandas as pd
model_dir = "models/"
model_names = sorted([name for name in os.listdir(model_dir) if os.path.isdir(os.path.join(model_dir, name))])

final = pd.DataFrame(columns=["Model", "Centre", 'Dice_LV_ED', 'Dice_MYO_ED', 'Dice_RV_ED', 'Dice_LV_ES', 'Dice_MYO_ES',
       'Dice_RV_ES'])

final[["Model", "Centre"]] = [[model, centre] for centre in range(6) for model in model_names]

for model in model_names:
    results = pd.read_csv(os.path.join(model_dir, model, "results.csv"))
    results = results.iloc[:, [1, 3, 4, 5, 7, 8, 9]]

    mean_results = results.groupby("Centre").mean()
    final.loc[final.Model==model, [ 'Dice_LV_ED', 'Dice_MYO_ED', 'Dice_RV_ED', 'Dice_LV_ES', 'Dice_MYO_ES',
       'Dice_RV_ES']] = mean_results.values
final.to_csv("results/final.csv", index=False)
```
